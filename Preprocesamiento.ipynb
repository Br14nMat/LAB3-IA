```python
import os
import cv2
import numpy as np
from PIL import Image, ImageEnhance
```

```python
target_size = (128, 128)  # Puedes ajustar este tamao segn tus necesidades

# Funcin para eliminar ruido usando el filtro GaussianBlur
def remove_noise(image):
    return cv2.GaussianBlur(image, (5, 5), 0)

# Funcin para ajustar el contraste usando Pillow
def adjust_contrast(image, factor=1.5):
    enhancer = ImageEnhance.Contrast(image)
    return enhancer.enhance(factor)

# Recorrer el directorio y preprocesar las imgenes
dataset_dir = 'dataset'
categories = ['perros', 'carros', 'frutas']

for category in categories:
    category_path = os.path.join(dataset_dir, category)
    for filename in os.listdir(category_path):
        if filename.endswith('.jpg') or filename.endswith('.png'):  # Ajusta segn tus tipos de archivo
            image_path = os.path.join(category_path, filename)
            
            # Leer la imagen usando OpenCV
            image = cv2.imread(image_path)
            
            # Eliminar ruido
            image_no_noise = remove_noise(image)
            
            # Convertir la imagen a PIL para ajustar el contraste
            image_pil = Image.fromarray(cv2.cvtColor(image_no_noise, cv2.COLOR_BGR2RGB))
            image_contrast = adjust_contrast(image_pil)
            
            # Redimensionar la imagen
            image_resized = image_contrast.resize(target_size)
            
            # Convertir la imagen de vuelta a formato OpenCV
            image_final = cv2.cvtColor(np.array(image_resized), cv2.COLOR_RGB2BGR)
            
            # Guardar la imagen preprocesada (puedes sobrescribir la original o guardar en un nuevo directorio)
            cv2.imwrite(image_path, image_final)
            
            print(f"Imagen preprocesada guardada: {image_path}")
```
